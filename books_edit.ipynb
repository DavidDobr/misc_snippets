```python
import zipfile

def unzip(file, targetdir):
    
    with zipfile.ZipFile(file,"r") as zip_ref:
        zip_ref.extractall(targetdir)
```

```python
import os
from os.path import join, getsize, getmtime
```

```python
# remember all files & timestamps
timestamps = list()

for root, dirs, files in os.walk('Strugatskie Vse Knigi/'):
    for f in files:
        # iterate over all files
        #print('='*50)
        # get full path to file
        if iszip(f):
            filepath = join(root,f)
            name = ''.join(f.split('.')[:-1]).lower()
            #print(name, "|", getmtime(archive))
            mtime =  datetime.datetime.fromtimestamp(getmtime(filepath))
            timestamps.append(tuple(
                                    [filepath, name, mtime]
                             ))

namedict = dict()
            
for num, value in enumerate(sorted(timestamps, key=lambda x: x[-1])):
    #print(str(num).zfill(2), value[1])
    namedict[value[1]] = str(num).zfill(2)
    
# each existing filename will get a number
#namedict['publ'] = '60'
namedict
```

```python
# remove all .diz and .zip
for root, dirs, files in os.walk('Strugatskie Vse Knigi/'):
    #print('-->',root, files)
    for f in files:
        filepath = join(root,f)
        name = ''.join(f.split('.')[:-1])
        #print(filepath, "|", name, "|", getmtime(filepath))
        if filepath[-4:].lower() in ['.diz', '.zip']:
            os.remove(filepath)
            print(filepath, ' removed!')
```

```python
# rename all .txt
for root, dirs, files in os.walk('Strugatskie Vse Knigi/'):
    #print('-->',root, files)
    for f in files:
        filepath = join(root,f)
        name = ''.join(f.split('.')[:-1]).lower()
        #print(filepath, "|", name, "|", getmtime(filepath))
        if filepath[-4:].lower() == '.txt':
            #print(filepath)
            
            try:
                #print(root+'/'+namedict[name]+ '_' + name)
                os.rename(filepath, dst=root+'/'+namedict[name]+ '_' + name)
            except KeyError:
                print("MISSING FILE {}".format(filepath))
```
